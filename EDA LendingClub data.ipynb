```
from pyspark.sql.functions import col,count,sum,when

display(df.groupby('addr_state').agg(count(col('annual_inc'))))
```

```
display(df.groupby('grade').agg(sum(col('loan_amnt'))))
```

```
display(df.groupby('grade','bad_loan').agg(sum('loan_amnt')))
```

```

df=df.withColumn('exposure',when(df.bad_loan=='No',col('revol_bal')).otherwise(-10*col('revol_bal')))
display(df)
```

```
display(df.groupby('bad_loan','grade').agg(sum('exposure')))
```
